```python
#=================================
# Load Dataset
#=================================

# Import dataset
df = pd.read_csv("music_data.csv")

# Display first 5 rows
print("First 5 rows of dataset:")
df.head()
```

**Insight:**
"The dataset is loaded successfully using pandas, and the first few rows are displayed to understand the structure and features of the data."

```python
#===============================
# Basic Information of Dataset
#===============================

# Shape of dataset
print("Shape of dataset:", df.shape)

# Dataset Structure
print("\nDataset Info:")
df.info()

# Statistical summary
print("\nStatistical Summary:")
print(df.describe())


```

**Insight:**
"The dataset overview provides basic structural information including the number of rows and columns, data types, and statistical summary. This helps in understanding the dataset size, feature types, and numerical distribution before further analysis."

```python
#==================
# Data Cleaning
#==================

# Missing values check
print("Missing Values in Each Column:\n")
print(df.isnull().sum())

# Duplicate records check
print("\nNumber of Duplicate Rows:")
print(df.duplicated().sum())
```

**Insight:**
"The dataset shows 0 missing values and 0 duplicate records, indicating that the data is already clean and well-structured. This ensures high data quality and allows reliable analysis without the need for additional cleaning steps."

```python
#=============================
# Emotion Distribution Analysis
#=============================

# Count emotions
emotion_counts = df["emotion"].value_counts()
print("Emotion Distributio:\n")
print(emotion_counts)

# Visualization
plt.figure(figsize=(8,5))

ax = emotion_counts.plot(
    kind="bar",
    color=["red","blue","green","orange","purple","brown","pink"],
    edgecolor="black"
)

# Titles and labels
plt.title("Emotion Distribution", fontsize=16, fontweight="bold")
plt.xlabel("Emotion", fontsize=14)
plt.ylabel("Count", fontsize=14)

# Improve readability
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.5)

# Final layout
plt.tight_layout()
plt.show()

```

**Insight:**
"This chart represent the frequency distribution of different emotions in the music dataset.
It helps us understand which types of emotional songs(motivational,romantic,sad,etc.) are more dominant in the dataset.

The emotion distribution shows that Motivational(38) is the most frequent emotion in the dataset, followed by Romantic(31) and Sad(30), indicating a strong presence of emotional and expressive content. Anger and Nostalgic emotions are equally represented(27 each), while Calm(25) and Happy(23) appear less frequently. Overall, the dataset is dominated by motivational themes rather than purely positive emotions."

```python
#================
# Rating Analysis
#================

# Basic statistics
avg_rating = df["rating"].mean()
min_rating = df["rating"].min()
max_rating = df["rating"].max()

print(f"Average Rating: {avg_rating:.2f}")
print(f"Minimum Rating: {min_rating:.2f}")
print(f"Maximum Rating: {max_rating:.2f}")

# Visualization
plt.figure(figsize=(8,5))


df["rating"].plot(
    kind="hist",
    bins=10,
    color="skyblue",
    edgecolor="black"
)

# Titles and labels
plt.title("Rating Distribution", fontsize=18, fontweight="bold")
plt.xlabel("Rating", fontsize=14)
plt.ylabel("Frequency", fontsize=14)

# Improve readability
plt.grid(axis="y", linestyle="--", alpha=0.5)

# Final layout
plt.tight_layout()
plt.show()
```

**Insight:**
"The rating analysis shows that the average rating is 3.06, indicating that most songs are of moderate quality. The minimum rating is 1.00 and the maximum rating is 5.00, which means the dataset contains both low-rated and highly rated songs. This suggests that while many entries are moderately rated, there are both very low and very high ratings, showing strong variability in user opinions."

```python
#===========================
# Genre Distribution Analysis
#===========================

# Count genres
genre_counts = df["genre"].value_counts()
print("Genre Distribution:\n")
print(genre_counts)

# Visualization
plt.figure(figsize=(8,5))

genre_counts.plot(
    kind="bar",
    color=plt.cm.Set3.colors,
    edgecolor="black"
)

# Titles and labels
plt.title("Genre Distribution", fontsize=18, fontweight="bold")
plt.xlabel("Genre", fontsize=14)
plt.ylabel("Count", fontsize=14)

# Improve readability
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.5)

# Final layout
plt.tight_layout()
plt.show()
```

**Insight:**
"The genre distribution shows that Rock(35), Pop(34), and HipHop(33) are the most dominant genres in the dataset, indicating high representation of modern and mainstream music. Indie(31) and Romantic(26) also have a significant presence, while Classsical(22) and Dance(20) are less represented compared to others. Overall, the dataset is slightly skewed towards contemporary genres."

```python
#==================================
# Language Distribution Analysis
#==================================

# Count languages
language_counts = df["language"].value_counts()
print("Language Distribution:\n")
print(language_counts)

# Visualization
plt.figure(figsize=(6,6))

language_counts.plot(
    kind="pie",
    autopct="%1.1f%%",
    colors=["#6A4C93","#1982C4","#8AC926","#FFCA3A","#FF595E","#00BFA6","#F72585"],
    wedgeprops={"edgecolor":"black"})

# Title
plt.title("Language Distribution", fontsize=14, fontweight="bold")

# Remove y-label
plt.ylabel("")

# Final layout
plt.tight_layout()
plt.show()
```

**Insight:**

The chart shows the Marathi and Hindi have the highest number of songs, followed by Punjabi and Telugu. Bengali, Tamil, and English have comparatively fewer songs. Overall, the distribution is quite balanced, with a slight dominance of Marathi and Hindi languages."

```python
#=================================
# Average Rating Trend Over Years
#=================================

# Calculate year-wise average rating
year_rating = df.groupby("year")["rating"].mean()
print(year_rating)

plt.figure(figsize=(10,5))

year_rating.plot(
    kind="line",
    marker="o",
    color="#1982C4"
)

plt.title("Average Rating Trend Over Years", fontsize=14, fontweight="bold")
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average Rating", fontsize=12)

plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

**Insight:**
"The average rating from 2000 to 2026 show a highly fluctuating pattern with no consistent upward or downward trend. Most ratings lie within a moderate range (2.5-3.5), but occasional sharp peaks and drops highlight periods of high and low performance, indicating variability satisfaction over time."

```python
#================================
# Year-wise Song Count Analysis
#================================

# Count songs per year(sorted)
year_count = df["year"].value_counts().sort_index()
print(year_count)

plt.figure(figsize=(8,5))

year_count.plot(
    kind="bar",
    color="#FF8C42",
    edgecolor="black"
)

plt.title("Number of Songs Released Each Year", fontsize=18, fontweight="bold")
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of Songs", fontsize=14)

plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

**Insight:**
"The chart shows the number of songs released each year from 2000 to 2026. The highest number of songs were released in 2017(17 songs), followed by 2019(12 songs) and 2005(11 songs), indicating peak production in these years. On the other hand, the lowest number of songs were recorded in 2023(2 songs) and 2011(3 songs), showing a decline in those years. Overall, the data reflects fluctuations in song releases over time, with no consistent upward or downward trend."

```python
#==========================
# Genre trend over years
#==========================

# Grouping data by year and genre
genre_trend = df.groupby(["year","genre"]).size().unstack(fill_value=0)
print(genre_trend)

# Display table
plt.figure(figsize=(10,6))

genre_trend.plot(
    kind="line",
    marker="o"
)

plt.title("Genre Trend Over Years", fontsize=14, fontweight="bold")
plt.xlabel("Year")
plt.ylabel("Number of Songs")

plt.legend(title="Genre", loc="upper left", bbox_to_anchor=(1,1))
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

**Insight:**
The chart shows the year-wise trend of different music genres, where each line represents a genre and its number of songs over time. The data indicates that genre popularity changes across years, with some genres increasing while other decrease. This highlights shifting audience preferences and evolving music trends over time."

```python
#====================
# Correlation
#====================

corr = df.corr(numeric_only=True)
print(corr)

plt.figure(figsize=(8,5))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5,
    cbar=True)

plt.title("Correlation Heatmap", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```

**Insight:**
"The correlation matrix shows that diagonal values are 1.00, which indicates perfect correlation of each variable with itself. The other values like 0.02, -0.00, and 0.07 are very close to zero, indicating very weak or no correlation between different variables. This means that features such as rating, year, and duration do not have a strong linear relationship and behave independently."

## SQL Analysis

```python
import sqlite3

conn = sqlite3.connect("music.db")
df.to_sql("music_data", conn, if_exists="replace", index=False)
```

```python
#=======================
# check table
# ======================

check = pd.read_sql("""
    SELECT * FROM music_data
    LIMIT 5
    """, conn)
print(check)
```

```python
#======================
# TOP RATED SONGS
#======================

top_song = pd.read_sql("""
SELECT title, artists, rating
FROM music_data
ORDER BY rating DESC
LIMIT 10
""", conn)
print(top_song)
```

```python
#=======================
# GENRE WISE SONG COUNT
#=======================

genre_count = pd.read_sql("""
SELECT genre, COUNT(*) AS total_songs
FROM music_data
GROUP BY genre
ORDER BY total_songs DESC
""", conn)

print(genre_count)
```

```python
#========================
# AVERAGE RATING BY GENRE
#========================

avg_rating = pd.read_sql("""
SELECT genre, AVG(rating) AS avg_rating
FROM music_data
GROUP BY genre
ORDER BY avg_rating DESC
""", conn)

print(avg_rating)
```

```python
#=========================
# SONGS RELEASED EACH YEAR
#=========================

release_year = pd.read_sql("""
SELECT year,COUNT(*) AS songs
FROM music_data
GROUP BY year
ORDER BY year
""", conn)

print(release_year)
```

```python
#========================
# LONGEST DURATION SONGS
#========================

longest_duration = pd.read_sql("""
SELECT title, artists, duration_min
FROM music_data
ORDER BY duration_min DESC
LIMIT 10
""", conn)

print(longest_duration)
```

```python
#========================
# EMOTION WISE SONGS
#========================

emotion_songs = pd.read_sql("""
SELECT emotion, COUNT(*) AS total_songs
FROM music_data
GROUP BY emotion
ORDER BY total_songs DESC
""", conn)

print(emotion_songs)
```
